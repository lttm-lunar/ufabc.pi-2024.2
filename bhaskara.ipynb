```python
import math
a = float(input())
b = float(input())
c = float(input())
x = float(input())

raiz1 = (-b + math.sqrt(b**2 - 4*a*c))/(2*a)
raiz2 = (-b - math.sqrt(b**2 - 4*a*c))/(2*a)

resultado = a*x**2+b*x+c

print(a)
print("raizes: {:.2f} {:.2f}".format(raiz1,raiz2))
print("resultado para x = {:.2f} {:.2f}".format(x,resultado))
```
