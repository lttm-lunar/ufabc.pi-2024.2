```python
import math
import numpy
import matplotlib.pyplot as plot

x= numpy.arange(0,2*math.pi,0.1)
print(x) #DOMINIO
f = numpy.sin(x)
print(f) #IMAGEM
plot.plot(x,f) #Gera o gráfico
plot.title('Teste') #Título

plot.xlabel('Eixo X') #Nomeação
plot.ylabel('Eixo Y') #Dos eixos
plot.grid(True, which='both') #Coloca grade
plot.plot(3,0,'ro') #Destaca o ponto 3,0
```
