```python
#questao 3
import numpy as np

redbull = 0
ferrari = 0
mercedes = 0
for i in range(1,23):
    print(str(i)+"a corrida:")
    r = np.random.rand() * 2
    f = np.random.rand() * 1.5
    m = np.random.rand()
    if (r > f and r > m):
        print("Red Bull venceu!")
        redbull += 25
        if (f > m):
            print("2o lugar: Ferrari")
            print("3o lugar: Mercedes")
            ferrari += 18
            mercedes += 15
        else:
            print("2o lugar: Mercedes")
            print("3o lugar: Ferrari")
            mercedes += 18
            ferrari += 15
    elif (f > r and f > m):
        print("Ferrari venceu!")
        ferrari += 25
        if (r > m):
            print("2o lugar: Red Bull")
            print("3o lugar: Mercedes")
            redbull += 18
            mercedes += 15
        else:
            print("2o lugar: Mercedes")
            print("3o lugar: Red Bull")
            mercedes += 18
            redbull += 15
    else:
        print("Mercedes venceu!")
        mercedes += 25
        if (r > f):
            print("2o lugar: Red Bull")
            print("3o lugar: Ferrari")
            redbull += 18
            ferrari += 15
        else:
            print("2o lugar: Ferrari")
            print("3o lugar: Red Bull")
            ferrari += 18
            redbull += 15
    print("------------------------------------------------")

if (redbull >= ferrari and redbull >= mercedes):
    print("CAMPEÃO: Red Bull com",redbull,"pontos")
    if (ferrari >= mercedes):
        print("2o lugar: Ferrari com",ferrari,"pontos")
        print("3o lugar: Mercedes com",mercedes,"pontos")
    else:
        print("2o lugar: Mercedes com",mercedes,"pontos")
        print("3o lugar: Ferrari com",ferrari,"pontos")
elif (ferrari >= redbull and ferrari >= mercedes):
    print("CAMPEÃO: Ferrari com",ferrari,"pontos")
    if (redbull >= mercedes):
        print("2o lugar: Red Bull com",redbull,"pontos")
        print("3o lugar: Mercedes com",mercedes,"pontos")
    else:
        print("2o lugar: Mercedes com",mercedes,"pontos")
        print("3o lugar: Red Bull com",redbull,"pontos")
else:
    print("CAMPEÃO: Mercedes com",mercedes,"pontos")
    if (redbull >= ferrari):
        print("2o lugar: Red Bull com",redbull,"pontos")
        print("3o lugar: Ferrari com",ferrari,"pontos")
    else:
        print("2o lugar: Ferrari com",ferrari,"pontos")
        print("3o lugar: Red Bull com",redbull,"pontos")
```

A) O programa funciona como um jogo, realizando simulações de 22 corridas com resultados aleatórios, e assim atribuindo pontos nos valores de "redbull", "mercedes" e "ferrari".

Um número aleatório  é selecionado e atribuído às variáveis "r", "f" e "m", que servem de base para saber quem ganhou a maior pontuação. O de maior valor recebe 25 pontos, o segundo colocado recebe 18, e o último 15; e em seguida são apresentados em um pódio, com título enumerando qual foi a corrida.

Por fim, a soma total dos pontos acumulados revelam quem foi o grande vencedor, colocando-os novamente em um pódio e mostrando as variáveis "r", "f" e "m" de acordo com suas marcas atribuídas; Red Bull, Ferrari e Mercedes, respectivamente.

B) Os resultados finais serão fortemente manipulados devido à multiplicação dos valores de "r" e "f", que lhes garantirão vantagem para alcançarem valores mais altos em relação a variável "m". Dessa forma, Red Bull e Ferrari terão maior probabilidade de alcançarem primeiro e segundo lugar, respectivamente, nos pódios finais.
