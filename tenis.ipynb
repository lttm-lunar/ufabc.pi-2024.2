```python
def tenis(a,u,v):
  if(a == "Nuvens"):
    return print("S")
  if(a == "Sol" and u == "Normal"):
    return print("S")
  if(a == "Sol" and u == "Elevada"):
    return print("N")
  elif(a == "Chuva" and v == "Fraco"):
    return print("S")
  elif(a == "Chuva" and v == "Forte"):
    return print("N")

a = str(input())
u = str(input())
v = str(input())

tenis(a,u,v)
```
